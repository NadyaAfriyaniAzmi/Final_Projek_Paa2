BUBBLE DAN INSERION SORT

```python
import time

def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    end_time = time.time()
    return arr, end_time - start_time

def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    end_time = time.time()
    return arr, end_time - start_time

def print_iterations(arr, sort_name, iterations):
    n = len(arr)
    print(f"\n{sort_name} - First {iterations} Iterations:")
    for i in range(iterations):
        print(arr[i], end=" ")
    print("\n")

    print(f"{sort_name} - Last {iterations} Iterations:")
    for i in range(n - iterations, n):
        print(arr[i], end=" ")
    print("\n")

def print_execution_time(sort_name, execution_time):
    print(f"{sort_name} - Execution Time: {execution_time:.8f} seconds")

def print_before_after(arr, sort_name):
    print(f"\n{sort_name} - Before Sorting: {arr}")
    sorted_arr, _ = bubble_sort(arr.copy()) if sort_name == "Bubble Sort" else insertion_sort(arr.copy())
    print(f"{sort_name} - After Sorting: {sorted_arr}")

def analyze_algorithm():
    arr = [
        12, 99, 62, 15, 20, 95, 39, 48, 3, 24, 8, 43, 74, 19, 97, 33, 49, 68, 55,
        33, 90, 90, 7, 26, 85, 46, 39, 40, 9, 36, 60, 64, 89, 31, 25, 71, 21, 23, 63, 84, 32, 5, 3, 44, 21,
        10, 21, 17, 50, 2, 36, 53, 79, 54, 19, 88, 1, 32, 31, 15, 6, 3, 1, 40, 22, 43, 18, 1, 77, 9, 59,
        40, 7, 41, 81
    ]

    while True:
        print("1. Bubble Sort")
        print("2. Insertion Sort")
        print("3. Keluar")

        choice = input("Pilih metode pengurutan (1/2/3): ")

        if choice == "1":
            sorted_arr, execution_time = bubble_sort(arr)
            print_iterations(sorted_arr, "Bubble Sort", 5)
            print_execution_time("Bubble Sort", execution_time)
            print_before_after(arr, "Bubble Sort")

        elif choice == "2":
            sorted_arr, execution_time = insertion_sort(arr)
            print_iterations(sorted_arr, "Insertion Sort", 5)
            print_execution_time("Insertion Sort", execution_time)
            print_before_after(arr, "Insertion Sort")

        elif choice == "3":
            print("Terima kasih!")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

analyze_algorithm()
```

```python
import time
import sys
from itertools import permutations

# Menghitung jarak antara dua titik
def calculate_distance(graph, path):
    distance = 0
    for i in range(len(path) - 1):
        if path[i] not in graph or path[i + 1] not in graph[path[i]]:
            return float('inf')
        distance += graph[path[i]][path[i + 1]]
    return distance

# Algoritma Travelling Salesman Problem (TSP) dengan pendekatan Brute Force
def tsp(graph, start):
    nodes = list(graph.keys())
    nodes.remove(start)
    min_distance = float('inf')
    min_path = []

    # Generate semua permutasi dari node
    for i, perm in enumerate(permutations(nodes)):
        path = [start] + list(perm) + [start]
        distance = calculate_distance(graph, path)
        if distance < min_distance:
            min_distance = distance
            min_path = path
        print_iteration(i + 1, path, distance)

    return min_path, min_distance

    # Generate semua permutasi dari node
    for perm in permutations(nodes):
        path = [start] + list(perm) + [start]
        distance = calculate_distance(graph, path)
        if distance < min_distance:
            min_distance = distance
            min_path = path
    return min_path, min_distance

# Algoritma Dijkstra untuk mencari jalur terpendek
def dijkstra(graph, start, end):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    visited = set()

    while True:
        min_node = None
        min_distance = float('inf')
        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_node = node
                min_distance = distances[node]

        if min_node is None:
            break

        visited.add(min_node)
        for neighbor in graph[min_node]:
            new_distance = distances[min_node] + graph[min_node][neighbor]
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    path = []
    current_node = end
    while current_node != start:
        path.append(current_node)
        current_node = min(graph[current_node], key=lambda x: distances[x])
    path.append(start)
    path.reverse()
    return path, distances[end]

# Mencetak hasil setiap iterasi
def print_iteration(iteration, path, distance):
    if distance != float('inf'):
        print(f"Iterasi {iteration}:")
        print("Path:", "->".join(path))
        print("Distance:", distance)
        print()
# Mencetak waktu komputasi perhitungan path
def print_execution_time(execution_time):
    print(f"Waktu Komputasi: {execution_time:.8f} detik")

# Mencetak hasil akhir (shortest path)
def print_shortest_path(path, distance):
    print("Hasil Akhir (Shortest Path):")
    print("Path:", "->".join(path))
    print("Distance:", distance)

# Melakukan analisis algoritma
def analyze_algorithm():
    graph = {
        'a': {'b': 12, 'c': 10, 'g': 12},
        'b': {'a': 12, 'c': 8, 'd': 12},
        'c': {'a': 10, 'b': 8, 'd': 11, 'e': 3, 'g': 9},
        'd': {'b': 12, 'c': 11, 'e': 11, 'f': 10},
        'e': {'c': 3, 'd': 11, 'f': 6, 'g': 7},
        'f': {'d': 10, 'e': 6, 'g': 9},
        'g': {'a': 12, 'c': 9, 'e': 7, 'f': 9}
    }

    while True:
        print("1. Travelling Salesman Problem (TSP)")
        print("2. Dijkstra")
        print("3. Keluar")

        choice = input("Pilih algoritma (1/2/3): ")

        if choice == "1":
            start = input("Masukkan titik awal: ")
            start = start.lower()
            if start not in graph:
                print("Titik awal tidak valid.")
                continue
            start_time = time.time()
            path, distance = tsp(graph, start)
            end_time = time.time()
            execution_time = end_time - start_time
            print_shortest_path(path, distance)
            print_execution_time(execution_time)

        elif choice == "2":
            start = input("Masukkan titik awal: ")
            start = start.lower()
            if start not in graph:
                print("Titik awal tidak valid.")
                continue
            end = input("Masukkan titik akhir: ")
            end = end.lower()
            if end not in graph:
                print("Titik akhir tidak valid.")
                continue
            start_time = time.time()
            path, distance = dijkstra(graph, start, end)
            end_time = time.time()
            execution_time = end_time - start_time
            print_shortest_path(path, distance)
            print_execution_time(execution_time)

        elif choice == "3":
            print("Terima kasih!")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

analyze_algorithm()
```
